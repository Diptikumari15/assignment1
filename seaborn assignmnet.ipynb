```python
Que 1: Name any five plots that we can plot using the Seaborn library. 
Also, state the uses of each plot.
```

```python
Answer: Seaborn is a data visualization library built on top of Matplotlib that provides
        a high-level interface for creating informative and attractive statistical graphics.
        Here are five common plots that can be created using Seaborn

       1. Scatter Plot:
          Use: Display the relationship between two numerical variables.
            
       2. Line Plot:
          Use: Show the trend of a numerical variable over a continuous interval or time.
            
       3. Bar Plot:
          Use: Display the distribution of a categorical variable or the relationship
               between two categorical variables.  
            
       4. Histogram:
          Use: Illustrate the distribution of a single numerical variable.   
           
       5. Box Plot:
          Use: Display the distribution of a numerical variable across different categories
               and identify potential outliers.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. 
Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
```

```python
Answer:
```

```python
import seaborn as sns
f=sns.load_dataset("fmri")
sns.lineplot(x='timepoint', y='signal',data=f, hue="event", style="region") 
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn.
Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
Answer:
```

```python
titanic_data=sns.load_dataset("titanic")
sns.boxplot(x="pclass",y="age",data=titanic_data)
```

```python
sns.boxplot(x="pclass",y="fare",data=titanic_data)
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the
'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
```

```python
Answer:
```

```python
import seaborn as sns
diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x='price', hue='cut')
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. 
Use the hue parameter for the "species" column of the iris dataset.
```

```python
Answer:
```

```python
import seaborn as sns
iris_data = sns.load_dataset("iris")
sns.pairplot(iris_data, hue='species')
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
Answer:
```

```python
import seaborn as sns
flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d', linewidths=.5)
```
